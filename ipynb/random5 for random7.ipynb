```python
import numpy as np
import matplotlib.pyplot as plt
a = []
for i in range(1000):
    a.append(np.random.randint(low=1,high=6)) # generator for random number 1-5
    
import seaborn as sns
sns.countplot(a)
plt.show() 
```

```python
b = []
for i in range(2000):
    num = (np.random.randint(low=1,high=6)-1)*5  # high digit
    num += np.random.randint(low=1,high=6)   # plus low digit
    if num > 21:   # skip 21-25
        continue
    if num%7 == 0:
        b.append(7)
    else:
        b.append(num%7)
plt.figure(dpi=100)
sns.countplot(b)
plt.show()   
```
