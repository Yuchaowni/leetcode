```python
def zeta(s,n):
    res = complex(0,0)
    yr = []
    yi = []
    for i in range(1,n+1):
        a = complex(pow(i,-s))
        res += a
        yr.append(res.real)
        yi.append(res.imag)
        #print(i,a,res)
    return res,yr,yi
```

```python
res, yr,yi = zeta(0.5+14.135j,100000)
res
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi=150)
plt.title("zeta function, s=0.5 + 14.135j")
plt.plot(yr,'b',label='real part')
plt.plot(yi,'r',label='imag part')
plt.legend()
plt.xlabel("n",fontsize=15)
plt.ylabel('$\zeta$',fontsize=15)
plt.grid()
plt.show()
```
